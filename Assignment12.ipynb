#1
```{r}
library(caret)

preProcess_missingdata_model

trainData <- predict(preProcess_missingdata_model, newdata = t)

```

#2
```{r}
library(caret)

myGrid = expand.grid(mtry = 2, splitrule = c("extratrees"),
                     min.node.size = c(1:3))

tm1 <- train(target~.,data = t, method = "ranger",
               trControl = trainControl(method ="cv", number = 7, verboseIter = TRUE),
               tuneGrid = myGrid)

tm1
```
