```{r}
library(ggplot2)
density_fun = function(x) 
{
  numerics <- sapply(x, is.numeric)
  x1 = x[, numerics]
  l = list()
  for (i in 1:length(x1)) {
    p=ggplot(x1, aes(x = x1[,i])) + geom_density()
    l[[i]] = p
  }
  return(l)
}
density_fun(x)
```

#7
```{r}
library(ggplot2)
bar_fun = function(x) 
{
  facts <- sapply(x, is.factor)
  x2 = x[, facts]
  l = list()
  for (i in 1:length(x2)) {
    p=ggplot(x2, aes(x = x2[,i])) + geom_bar()
    l[[i]] = p
  }
  return(l)
}
bar_fun(x)
```
